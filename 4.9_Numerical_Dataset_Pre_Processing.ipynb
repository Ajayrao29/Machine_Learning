Importing the dependencies (libraries/functions used for our project)

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

Data Collection & Preprocessing

```python
#loading the data from csv file to a pandas dataframe
diabetes_dataset = pd.read_csv('diabetes.csv')
```

Separating features (first 8 columns in our dataset) and target(outcome column)

```python
X=diabetes_dataset.drop(columns='Outcome',axis=1)
#AXIS=1 if we want Column to be dropped and AXIS=0 if we want ROW to be dropped
Y=diabetes_dataset['Outcome']

#dividing features and target into individual sets
```

```python
print(X) #features set (outcome(9th column) not included)
```

```python
print(Y) #outcome set (features(8 columsn) not included)
```

Data Standardisation

Standardise data i.e; bring all values in the dataset to a common range inorder to get better and accurate predictions from ML model and its also make ML model to learn easily

--> Standardise dataset before splitting into training and testing because it retains original range of datset

-->from sklearn.preprocessing import StandardScaler

included as it helps in standardising

```python
standardised_data =StandardScaler().fit_transform(X) # STANDARDISING
#makes complete data to a common range
print(standardised_data)
```

```python
X=standardised_data
print(X)
print(Y)
```

Splitting the dataset into training and testing data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
#test_size=0.2 ,it means 20% data.
#random state is way of splitting data, will differ if it is equal to 3
```
